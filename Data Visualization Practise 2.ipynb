```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
%matplotlib inline
```

```python
apl_price = [93.95,112.15,104.05,144.85,169.49]
ms_price = [39.01,50.29,57.05,69.98,94.39]
year = [2014,2015,2016,2017,2018]
```

```python
# Set the plot axies with year, apl_price while customizing them with 'k' for black color and 'g' for green
plt.plot(year, apl_price,'*k',year,ms_price,'g')
plt.xlabel('Year')
plt.ylabel('Stock Price')
#plt.axis([2013,2019,20,180]) => Set the maximum and minimum of the axies
plt.show()
```

```python
# Figure is the container of plots
# 6.4, 8.4 are default values for the figure
fig_l = plt.figure(1,figsize=(20,4.8))
# The number in add_subplot() shows you how many charts you want to display in yopur figure and how the chart is positioned inside
# structured. 121 means 1 row - 2 columns on the 1st poistion while 122 means 1 row - 2 columns on the 2nd position
apple_price = fig_l.add_subplot(121)
microsoft_price = fig_l.add_subplot(122)

apple_price.plot(year,apl_price)
# Set the axises to round up number
apple_price.xaxis.set_major_locator(MaxNLocator(integer=True))

microsoft_price.scatter(year,ms_price)

# Set the axises to round up number
microsoft_price.xaxis.set_major_locator(MaxNLocator(integer=True))

# By using show(), the line of memory is removed
plt.show()
```

```python
# Create a figure with 2 rows, 2 columns and sizes
fig_2, axes = plt.subplots(2,2,figsize=(20,5))
# Set the plot position. If the figure has 1 columns, 2 rows, then we will use axes[0] and axes[1] instead of axes[0,1] axes[1,0]
axes[0,1].scatter(year,ms_price)
axes[1,0].plot(year,apl_price)
plt.show()
```
