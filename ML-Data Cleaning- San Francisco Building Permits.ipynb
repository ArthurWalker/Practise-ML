```python
# Calculate total number of cells
totalCells = np.product(sfPermits.shape)
sfPermits.shape
```

```python
#Count the number of missing values per column
missingCount = sfPermits.isnull().sum()

#Calculate total number of misisng values
totalMissing = missingCount.sum()
```

```python
print (missingCount)
print ('Total Missing:',totalMissing)
print ('THe SF Permits dataset contains', round(((totalMissing/totalCells)*100),2),'%','missing values')
```

```python
sfPermitsCleanCols=sfPermits.dropna(axis=1)
print ('Columns in original dataset',sfPermits.shape[1])
print ('Columns with na dropped',sfPermitsCleanCols.shape[1])
```

```python
# Fill the missing data which is below 1 line exactly the same as the previous line
sfPermits['Proposed Construction Type'].head().fillna(method='pad',limit=1,inplace=True)
```

```python
# Fill the missing data which is above 1 line exactly the same as the below line
sfPermits['Proposed Construction Type'].head().fillna(method='bfill',limit=1,inplace=True)
```
